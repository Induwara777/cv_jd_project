# Names Function

```python
def cv_names(folder):
    names = [os.path.basename(os.path.join(folder,f)) for f in os.listdir(folder) if f.lower().endswith(".pdf")]
    return names
```

# Save JSON file

```python
def cv_json(json_file,folder,x):
    with open(os.path.join(folder,f"{x}.json"),"w") as f:
        json.dump(json_file,f,indent=4)
```

```python
import tiktoken

def token_count(text):
    encoding = tiktoken.get_encoding("cl100k_base")
    return len(encoding.encode(text))
```
